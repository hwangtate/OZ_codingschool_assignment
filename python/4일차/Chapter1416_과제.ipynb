[문제4]

행렬의 덧셈은 행과 열의 크기가 같은 두 행렬의 같은 행, 같은 열의 값을 서로 더한 결과가 됩니다. 2개의 행렬 arr1과 arr2가 주어졌을 때, 행렬 덧셈의 결과를 출력하는 코드를 작성해주세요(result와 같은 결과가 출력되도록 작성)

arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = [[4,6],[7,9]]

```python
#정답 봐버린 문제

arr1 = [[1,2],[2,3]]
arr2 = [[3,4],[5,6]]

result = [] #리스트 none값 주기
for i in range(len(arr1)): #2번 반복하라 i = 0, i = 1
    row = [] #리스트 none값 주기 , 행이라 초기화 작업
    for j in range(len(arr1[0])): #2번 반복 j = 0, j =1
        row.append(arr1[i][j] + arr2[i][j])#각 리스트 마다 첫 번쨰, 두 번쨰 열 꺼내서 더해서 행으로 만들기
    result.append(row) #열 더해서 행으로 만든 값 결과에 넣어주기

print(result) 
```

0부터 9까지의 숫자 중 일부가 들어있는 정수 배열 numbers가 주어집니다.   
numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 출력해주세요

numbers = [5,8,1,0,6,9]

```python
numbers = [5,8,1,0,6,9]

find_numbers = list(range(0,10))

numbers.extend(find_numbers)

result = 0

for i in set(numbers):
    result += i

print(result)
```

[문제 2: 붕어빵 주문 받기]

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

```python
주문_붕어빵_종류, 주문_붕어빵_개수 = input("종류와 개수를 콤마로 구분해서 입력해주세요:").split(',')

주문내역 = {}

주문내역[주문_붕어빵_종류] = 주문_붕어빵_개수

print(f"주문하신 내용은 : {주문내역}입니다.")
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python

if 붕어빵_재고[주문_붕어빵_종류] >= int(주문내역[주문_붕어빵_종류]):
    붕어빵_재고[주문_붕어빵_종류] -= int(주문내역[주문_붕어빵_종류])
elif 붕어빵_재고[주문_붕어빵_종류] < int(주문내역[주문_붕어빵_종류]):
    print("재고가 부족합니다.")

print(붕어빵_재고)
print(주문내역)
```

[문제 4: 여러 손님 주문 처리]

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
while True:
    주문_붕어빵_종류 = input("붕어빵 종류를 입력해주세요 :")
    if 주문_붕어빵_종류 == "종료":
        break

    주문_붕어빵_개수 = input("붕어빵 개수를 입력해주세요 :")
    if 주문_붕어빵_개수 == "종료":
        break

    주문내역[주문_붕어빵_종류] = 주문_붕어빵_개수 #같은 종류의 붕어빵 개수 더해서 주문내역에 넣는 기능구현은 나중에 하겠습니다! 

    if 붕어빵_재고[주문_붕어빵_종류] >= int(주문내역[주문_붕어빵_종류]):
        붕어빵_재고[주문_붕어빵_종류] -= int(주문내역[주문_붕어빵_종류])
    elif 붕어빵_재고[주문_붕어빵_종류] < int(주문내역[주문_붕어빵_종류]):
        print("재고가 부족합니다.")
        
    print(f"주문하신 내용은 : {주문내역}입니다.")
    print(f"현재 재고는 : {붕어빵_재고}입니다.")
```

[문제 5: 관리자 모드]

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
print("<관리자 모드>")
while True:
    주문_붕어빵_종류 = input("추가할 붕어빵 종류를 입력해주세요 :")
    if 주문_붕어빵_종류 == "종료":
        break

    주문_붕어빵_개수 = input("추가할 붕어빵 개수를 입력해주세요 :")
    if 주문_붕어빵_개수 == "종료":
        break

    재고추가내역 = {
        주문_붕어빵_종류 : 주문_붕어빵_개수
    }

    붕어빵_재고[주문_붕어빵_종류] += int(재고추가내역[주문_붕어빵_종류])

    print(붕어빵_재고)
```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
가격 = {
    "팥 붕어빵": 1000,
    "슈크림 붕어빵": 1200,
    "초코 붕어빵": 1500
}

판매개수 = {
    "팥 붕어빵": 0,
    "슈크림 붕어빵": 0,
    "초코 붕어빵": 0
}

매출 = {
    "팥 붕어빵": 0,
    "슈크림 붕어빵": 0,
    "초코 붕어빵": 0
}

for key in 주문내역:
    판매개수[key] += int(주문내역[key]) 

print(f"주문내역 : {주문내역}")

총매출 = 0

for j in 매출:
    매출[j] = 가격[j] * 판매개수[j]

    총매출 += 매출[j]


print(f"매출 : {매출}") 
print(f"총매출 : {총매출}원")

#쉽네요 더 어렵게 내주세요(장난입니다)
#같은 종류의 붕어빵을 여러번 주문했을때 주문개수를 더해서 주문내역에 넣고 매출까지 계산해주는 코드는 구현 못함 
```
