파이썬은 변수를 할당 할때 변수 그 자체에 그 값을 지정 하는 것이 아닌 담고 있는 것이다
예를들어, a = 1 이라고 한다면 a는 1이라는 값의 위치를 담고 있는 것이다 
그래서 b = a 라고 한다면 사실상 b = 1이다 왜냐면 a는 그저 값의 위치를 저장하고 있기 떄문이다

하지만 다른 언어는 a = 1 이라고 한다면 실제로 a라는 변수에 1이라는 값을 저장하는 것이다
그렇기 때문에 b = a 라고 한다면 이건 그대로 b = a라고 해야한다
